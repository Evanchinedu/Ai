```python
# Guess Game
import random
score = 0

live = 5  
player =  input ('Enter your name: ')
print (f' hi, {player}. Welcome to playstation!')
while True:
    if live ==0:
       print ('Game over!')
       break
systemGuess = random.randint(1,6)
myGuess = int (input('Guess a number(1-6) : '))
if systemGuess == myGuess:
   print (f'Great! System guessed {systemGuess}')
   score = score + 5
   print (f'your new score is {score}')
else:
   print (f'sorry! system guessed {systemGuess}')
   score = score-5
   live = live-1
   print (f'your new score is {score} and live left is {live}')
```

```python
# Contact Book Program

contacts = {}

while True:
    print("\nContact Book")
    print("1. Add Contact")
    print("2. Search Contact")
    print("3. Quit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        name = input("Enter name: ")
        phone = input("Enter phone number: ")
        contacts[name] = phone
        print(f"{name} added successfully!")

    elif choice == "2":
        name = input("Enter name to search: ")
        if name in contacts:
            print(f"{name}'s phone number is {contacts[name]}")
        else:
            print("Contact not found.")

    elif choice == "3":
        print("Exiting Contact Book. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
